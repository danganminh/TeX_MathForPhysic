```python
import numpy as np
import matplotlib.pyplot as plt
```

# Beginner Application of Differentiation

```python
# ex 9, page 34/69

def f(x):
    return 3*x**2 - 7*x + 5

def gradient_f(x):
    return 6*x - 7

x = np.linspace(-2, 3, 100)
y = f(x)
gradient_y = gradient_f(x)

plt.plot(x, y, label=r"$y = 3x^2 - 7x + 5$")
plt.plot([1, 1], [-1, 0], color="red", label="(1,1)")
plt.plot(x, gradient_y, label=r"$y'$")
plt.plot([-2, 2], [0, 0], linestyle="--", color='grey')

plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-2, 3)
plt.xlim(0, 2)
plt.legend()
```
